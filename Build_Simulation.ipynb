```python
import random
```

```python
#All Variables to consider
startPop=50 #Starting the population with 50 people
infantMortality = 5 #infant mortality of 5%
youthMortality = 45 
agriculture = 5 #How many units of food each person generates
disasterChance = 10 
fertilityx = 18 #Start of reproductive age for women
fertilityy = 35 #End of reproductive age for women
food = 0 #Food reserves
peopleDictionary = []
```

```python
class Person:
  def __init__(self,age):
    self.gender=random.randint(0,1)
    self.age=age
    self.pregnant=0

def beginSim():
  '''Start Simulation'''
  for x in range(startPop):
    peopleDictionary.append(Person(random.randint(18,50)))

def harvest(food,agriculture):
  '''Harvest'''
  ablePeople=0
  for person in peopleDictionary:
    if person.age>8: #Anybody above the age of 8 is considered to be able bodied
      ablePeople +=1

  food +=ablePeople*agriculture
  if food < len(peopleDictionary): #If food reserves is less than the number of people alive death occurs
    del peopleDictionary[0:int(len(peopleDictionary)-food)]
    food=0

  else:
    food -=len(peopleDictionary) 

def reproduce(fertilityx,fertilityy,infantMortality):
  '''Reproduction'''
  for person in peopleDictionary:
    if person.gender ==1: #Assuming 1 denotes female and 0 denotes male
      if person.age > fertilityx and person.age < fertilityy:
        if random.randint(0,5)==1: # One in 5 women is able to give birth to a healthy baby
          if random.randint(0,100)>infantMortality: #95% chance infant mortality doesn't take place
            peopleDictionary.append(Person(0))

pop=[]
def runYear(food, agriculture, fertilityx, fertilityy,infantMortality, disasterChance):
  '''Count Year'''
  harvest(food, agriculture)
  reproduce(fertilityx, fertilityy,infantMortality)
  for person in peopleDictionary:
    if person.age>80: #Life expetancy is set to 80 years
      peopleDictionary.remove(person)

    else:
      person.age +=1
  pop.append(len(peopleDictionary))
  print(len(peopleDictionary))
  infantMortality *=0.985
  return infantMortality
  
#If we get a range between the disasterchance, some disaster happens and kills 5-20% of the population
  if random.randint(0,100)<disasterChance:
        del peopleDictionary[0:int(random.uniform(0.05,0.2)*len(peopleDictionary))]


beginSim()
while len(peopleDictionary)<100000 and len(peopleDictionary)>1:
  infantMortality=runYear(food, agriculture, fertilityx, fertilityy,infantMortality,disasterChance)
```
